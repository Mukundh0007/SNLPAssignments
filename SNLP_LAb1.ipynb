```python
import wikipedia
wiki = wikipedia.page('Vinícius Júnior')
text = wiki.content
text = text.replace('==', '')
print(text)
```

# Tokenization

```python
from nltk.tokenize import sent_tokenize, word_tokenize
sentence_tokens=sent_tokenize(text)
word_tokens=word_tokenize(text)
print(sentence_tokens)
print(word_tokens)
```

Tokenization:
Tokenization is the process of breaking down a text into smaller units called tokens.

These tokens can be sentences (sentence tokenization) or words (word tokenization).

In the code above, `sent_tokenize` splits the text into sentences, and `word_tokenize` further splits the text into individual words.


```python
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+') # only words no spl characters
regexp_tokens = tokenizer.tokenize(text)
print(regexp_tokens)
```

The RegexpTokenizer is used to tokenize text based on a regular expression pattern.
In this case, the pattern r'\w+' matches one or more word characters (alphanumeric characters plus underscore).
Therefore, the `regexp_tokens` variable will contain a list of all the words in the text,
excluding punctuation and other non-word characters.

```python
from nltk.tokenize import WordPunctTokenizer
word_punct_tokenizer = WordPunctTokenizer()
word_punct_tokens = word_punct_tokenizer.tokenize(text)
print(word_punct_tokens)
```

 The WordPunctTokenizer splits the text into words and punctuation marks as separate tokens.
This means that punctuation marks are treated as individual tokens, unlike in the RegexpTokenizer example where they were excluded.

For example,

# Stemming

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()

for w in word_tokens:
    print(w,':',ps.stem(w))
```

```python
from nltk.stem import LancasterStemmer
ls = LancasterStemmer()
for w in word_tokens:
    print(w,':',ls.stem(w))
```

Stemming differences:
1. "commonly" is stemmed to "commonli" by PorterStemmer, but to "common" by LancasterStemmer.
   This shows that LancasterStemmer is more aggressive in removing suffixes.
2. "national" is stemmed to "nation" by PorterStemmer, but to "nat" by LancasterStemmer.
   Again, LancasterStemmer removes more of the word.

Therefore,
- PorterStemmer uses a set of rules to iteratively remove suffixes from words. It is widely used and considered a good balance of accuracy and efficiency.
- LancasterStemmer is a more aggressive stemming algorithm that often removes more of the word than PorterStemmer. It can be useful for tasks where reducing the vocabulary size is important.


```python
from nltk.stem import SnowballStemmer
ss = SnowballStemmer("english")
for w in word_tokens:
    print(w,':',ss.stem(w))
```

```python

unique_stems = []
for w in word_tokens:
  ps_stem = ps.stem(w)
  ls_stem = ls.stem(w)
  ss_stem = ss.stem(w)
  if not (ps_stem == ls_stem and ls_stem == ss_stem):
    unique_stems.append((w, ps_stem, ls_stem, ss_stem))

print(unique_stems)
```

Snowball Stemmer is based on a more extensive set of rules and language-specific knowledge compared to Porter.
It aims to produce more accurate and linguistically sound stems, leading to better handling of irregular verbs and other edge cases.
This explains why Snowball Stemmer often produces more meaningful stems than Porter or Lancaster in certain cases.

# Lemmatization

```python

nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
for w in word_tokens:
    print(w,':',lemmatizer.lemmatize(w))
```

# Comparison



Stemming:
- Reduces words to their root form by removing suffixes.
- May not always produce valid words (e.g., "playing" -> "play").
- Generally faster than lemmatization.

Lemmatization:
- Reduces words to their base or dictionary form (lemma).
- Considers the context of the word to produce valid words.
- Usually slower than stemming.

In the provided code:
- Stemming: Words like "playing" are stemmed to "play," which is a valid word.
- Lemmatization: Words like "playing" are lemmatized to "play," considering the context.

Choice depends on the application:
- Stemming: Suitable for tasks like basic search or text classification where exact word forms are less critical.
- Lemmatization: Preferred for tasks like natural language understanding or machine translation where accurate word forms are essential.
